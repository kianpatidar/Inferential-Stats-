#**Statistical Analysis**

```python
# -->Data collection
# -->preparing samples
# -->creating two hypothesis (null hypothesis,alternate hypothesis)
# -->applying the appropriate test(based on requirement or case)
# -->if p-value>0.05 accept null hypothesis
#    if p-value<0.05 Rejection of null hypothesis
```

#**a. It was observed that the average purchase made by the Men of the age 18-25 was 10000. Is it still the same?**

```python
new_data = data[(data["Age"]==1) & (data["Gender"]==1)]
```

```python
new_data
```

ONE SAMPLE T-test

```python
# null hypothesis: means will be equal to 10000
# alternate hypothesis : means will not be equal to 10000
```

```python
sample = new_data.sample(1000,random_state=5)
```

```python
sample
```

```python
pop_mean = 10000
```

```python
from scipy.stats import ttest_1samp
```

```python
sample["Purchase"].mean()
```

```python
t_stat,p_value = ttest_1samp(sample["Purchase"],pop_mean)
```

```python
print(t_stat,p_value)
```

```python
# since the p_value is less than 0.05, we will reject the null hypothsesis and accept the alternate hypothesis
```

#**b. It was observed that the percentage of women who spend more than 10000 was 35%. Is it still the same?**

```python
data_new = data.loc[(data["Purchase"]>10000)]
```

```python
data_new
```

```python
#no of women in the data

count = data_new["Gender"].value_counts()[0]
```

```python
count
```

```python
#number of observations

nobs = len(data_new["Gender"])
```

```python
nobs
```

```python
p0 = 0.35
```

```python
(10175/49390)*100
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
z_stat, p_value = proportions_ztest(count = count, nobs = nobs, value=p0)
```

```python
print(z_stat, p_value)
```

```python
#null hypothesis - proportion is 35%.
#alternate hypothesis - proportion is not 35%.
```

```python
# since the p_value is less than 0.05, we will reject the null hypothseis and accept the alternate hypothesis
```

#**c. Is the average purchase made by men and women of the age 18-25 same**

TWO SAMPLE Ttest: comparing the mean of two sample given that the variance of both the samples are equal by default

```python
# h0 - average purchase are equal for male and female
# h1 - average purchase are not equal for male and female
```

```python
data_men = data.loc[(data["Gender"]==1) & (data["Age"] == 1)]
data_women = data.loc[(data["Gender"]==0) & (data["Age"] == 1)]
```

```python
data_men
```

```python
data_women
```

```python
#sample
```

```python
data_men_sample = data_men.sample(500,random_state=0)
data_women_sample = data_women.sample(500,random_state=0)
```

```python
data_men_sample
```

```python
data_women_sample
```

```python
from scipy.stats import ttest_ind
```

```python
data_men_sample.Purchase.var()
```

```python
data_women_sample.Purchase.var()
```

```python
t_stat,p_value = ttest_ind(data_men_sample["Purchase"],data_women_sample["Purchase"])
```

```python
print(t_stat,p_value)
```

```python
#we can reject the null hypothesis
```
